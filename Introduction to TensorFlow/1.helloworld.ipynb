```python
import tensorflow as tf
import numpy as np
print(tf.__version__)
```

```python
x=np.array([-1.0,0.0,1.0,2.0,3.0,4.0])
y=np.array([-3.0,-1.0,1.0,3.0,5.0,7.0])
model=tf.keras.Sequential([tf.keras.layers.Dense(units=1,input_shape=(1,),name = 'layer1')])
model.compile(optimizer=tf.keras.optimizers.SGD(),loss=tf.keras.losses.mean_squared_error)
model.fit(x,y,epochs=2000)
```

```python
W1, b1 = model.get_layer("layer1").get_weights()
print(f"W1{W1.shape}:\n", W1, f"\nb1{b1.shape}:", b1)
print("Benar karena rumusnya memang 2x-1")
```
